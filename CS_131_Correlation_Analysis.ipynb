```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/cleaned_students_mental_health_survey.csv')
df
```

```python
# Create a mapping for 'Sleep_Quality'
sleep_quality_mapping = {'Poor': 1, 'Average': 2, 'Good': 3}

# Replace categorical values with numerical representations
df['Sleep_Quality'] = df['Sleep_Quality'].map(sleep_quality_mapping)

# Calculate the correlation matrix
correlation_matrix = df[['Stress_Level', 'Sleep_Quality', 'Depression_Score', 'Anxiety_Score']].corr()
print(correlation_matrix)
```

```python
# Display color-coded heatmap of the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Pearson Correlation Heatmap")
plt.show()
```
